```python
import pandas as pd
import mysql.connector


def clean_column_names(df):
    df.columns = df.columns.str.strip().str.lower()  
    return df


def limit_rows(df):

    if len(df) > 100:
        return df.head(50)  
    elif len(df) > 50:

        return df.head(20)
    else:
        return df  


with mysql.connector.connect(
    host='localhost',
    port=3307,
    user='root',
    password='root',
    database='WaterManagement'
) as connection:

    program_types_df = pd.read_sql("SELECT * FROM `ProgramTypes`;", connection)
    water_conservation_programs_df = pd.read_sql(
        "SELECT * FROM `WaterConservationPrograms`;", connection)
    water_source_df = pd.read_sql("SELECT * FROM `WaterSource`;", connection)
    departments_df = pd.read_sql("SELECT * FROM `Departments`;", connection)
    water_usage_records_df = pd.read_sql(
        "SELECT * FROM `WaterUsageRecords`;", connection)
    treatment_equipment_df = pd.read_sql(
        "SELECT * FROM `TreatmentEquipment`;", connection)
    maintenance_logs_df = pd.read_sql(
        "SELECT * FROM `MaintenanceLogs`;", connection)
    water_quality_tests_df = pd.read_sql(
        "SELECT * FROM `WaterQualityTests`;", connection)
    customer_details_df = pd.read_sql(
        "SELECT * FROM `CustomerDetails`;", connection)
    customer_conservation_programs_df = pd.read_sql(
        "SELECT * FROM `CustomerConservationPrograms`;", connection)


program_types_df = clean_column_names(program_types_df)
water_conservation_programs_df = clean_column_names(
    water_conservation_programs_df)
water_source_df = clean_column_names(water_source_df)
departments_df = clean_column_names(departments_df)
water_usage_records_df = clean_column_names(water_usage_records_df)
treatment_equipment_df = clean_column_names(treatment_equipment_df)
maintenance_logs_df = clean_column_names(maintenance_logs_df)
water_quality_tests_df = clean_column_names(water_quality_tests_df)
customer_details_df = clean_column_names(customer_details_df)
customer_conservation_programs_df = clean_column_names(
    customer_conservation_programs_df)


program_types_df = limit_rows(program_types_df)
water_conservation_programs_df = limit_rows(water_conservation_programs_df)
water_source_df = limit_rows(water_source_df)
departments_df = limit_rows(departments_df)
water_usage_records_df = limit_rows(water_usage_records_df)
treatment_equipment_df = limit_rows(treatment_equipment_df)
maintenance_logs_df = limit_rows(maintenance_logs_df)
water_quality_tests_df = limit_rows(water_quality_tests_df)
customer_details_df = limit_rows(customer_details_df)
customer_conservation_programs_df = limit_rows(
    customer_conservation_programs_df)


merged_df = pd.merge(water_conservation_programs_df,
                     program_types_df, on='programtypeid', how='inner')
print("After merging WaterConservationPrograms and ProgramTypes:")
print(merged_df.head())

merged_df = pd.merge(merged_df, water_usage_records_df,
                     left_on='programid', right_on='departmentid', how='inner')
print("After merging with WaterUsageRecords:")
print(merged_df.head())

merged_df = pd.merge(merged_df, departments_df, on='sourceid', how='inner')
print("After merging with Departments:")
print(merged_df.head())

merged_df = pd.merge(merged_df, water_source_df, on='sourceid', how='inner')
print("After merging with WaterSource:")
print(merged_df.head())

merged_df = pd.merge(merged_df, treatment_equipment_df,
                     on='sourceid', how='inner')
print("After merging with TreatmentEquipment:")
print(merged_df.head())

merged_df = pd.merge(merged_df, maintenance_logs_df,
                     on='equipmentid', how='inner')
print("After merging with MaintenanceLogs:")
print(merged_df.head())

merged_df = pd.merge(merged_df, water_quality_tests_df,
                     on='sourceid', how='inner')
print("After merging with WaterQualityTests:")
print(merged_df.head())

merged_df = pd.merge(merged_df, customer_conservation_programs_df,
                     left_on='programid', right_on='programid', how='inner')
print("After merging with CustomerConservationPrograms:")
print(merged_df.head())

merged_df = pd.merge(merged_df, customer_details_df,
                     left_on='customerid', right_on='customerid', how='inner')
print("After merging with CustomerDetails:")
print(merged_df.head())


merged_df.to_csv('mysql_merged.csv', index=False)
print("Merged DataFrame has been saved as 'mysql_merged.csv'")
```
